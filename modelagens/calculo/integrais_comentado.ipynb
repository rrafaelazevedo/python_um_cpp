# Cálculo de **Integrais Definidas**

```python
# Parâmetros necessários para o cálculo debaixo da curva 'f(x)'

# 'f(x)'
f = lambda x: x ** 2

# limite inferior
a = -10

# limite superior
b = 10

# altura do trapézio
dx = 1e-1 # é o mesmo que 1*10 ** (-1) ou 0.1
```

```python
# 'x' é a variável criada para percorrer o intervalo de 'a' até 'b'
x = a
areas_trapezios = []
while x <= b:
    # sequencia de prints para entender o que está acontecendo
    # print(x)
    # print(f(x))
    # print(f(x + dx))

    # altura da curva 'f(x)' nos pontos 'x' e 'x + dx'. São bases de um trapézio
    base_um = f(x)
    base_dois = f(x + dx)

    # print(base_um, base_dois)

    # para cada iteração, calcula-se a área de um trapézio e armazena-se em uma lista
    area_trapezio = (base_um + base_dois) * dx / 2
    # print(area_trapezio)
    areas_trapezios.append(area_trapezio)

    # incrementa 'x' para percorrer o intervalo de 'a' até 'b'
    x = x + dx

    # print("-" * 100)

# no final do loop, imprime a lista com as áreas dos trapézios de cada iteração
print(areas_trapezios)
```

```python
# somando a lista de áreas dos trapézios, tem-se a área total debaixo da curva 'f(x)' no intervalo de 'a' até 'b'
sum(areas_trapezios)
```

```python
from typing import Callable
def integrais_definidas(f: Callable, a: float | int, b: float | int, dx: float = 1e-3):

    """
    Essa função calcula a área de baixa da curva 'f' limitada entre
    os limites inferior (a) e superior (b) com precisão (dx) pelo método
    dos trapézios.
    """

    if not callable(f):
        raise Exception("O parâmetro 'f' deve ser do tipo 'function'.")

    if a > b:
        a, b = b, a

    # ----- LÓGICA DE PROGRAMAÇÃO ------
    x = a
    areas_trapezios = []
    while x <= b:
        base_um = f(x)
        base_dois = f(x + dx)
        area_trapezio = (base_um + base_dois) * dx / 2
        areas_trapezios.append(area_trapezio)
        x = x + dx
    # ----------------------------------

    return round(sum(areas_trapezios), ndigits=2)
```

```python
integrais_definidas(f=lambda x: x ** 3, a=2, b=4)
```
